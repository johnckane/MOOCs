Consider a Dataframe named df that contains 100 rows of data in 20 columns. Enter the line of code to return only the values in row five using the sub() function.

```julia
a1 = Array(randn(20));
a2 = Array(randn(100));
m1 = a1*a2';
df = convert(DataFrame,m1);
sub(df,5)

```

Write the line of code to permanently delete rows three, five, nine from a DataFrame with the name df.

```julia
df = DataFrame(a = 1:10,b=2:11,c=3:12);
deleterows!(df,[3,5,9])
```
