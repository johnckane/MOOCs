# Week 3: Fitting A Curve

```julia
# Loading the Data
using DataFrames
data_tofit = readtable("/home/john/MOOCs/Computer Science/Julia/Week3/Week3_PR_Data.dat",separator = ' ')
```

```julia
x = data_tofit[:,1]
y = data_tofit[:,2]
```

```julia
# Plot the data
using Plots
plot(x,y)
```

Define a function to fit the data

```julia
f(x) = a*x^2 + b*x + c
```

```julia
a,b,c = 1,1,1
plot(f,-5,5)
```

```julia
# Experiment with parameters to fit the data
a,b,c = -0.5,-0.5,6
plot(x,y)
plot!(f,0,6)
```
